```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
movies_data = pd.read_csv('movies_complete.csv')
movies_data.head()
```

```python
movies_data['non_dom_gross_usd'] = movies_data['worldwide_gross_usd'] - movies_data['domestic_gross_usd']
movies_data['title_length'] = movies_data['movie_title'].apply(lambda x: len(x))
pd.set_option('display.max_columns', None)
movies_data.head()
```

```python
movies_data['worldwide_roi'] = movies_data['worldwide_gross_usd'] / movies_data['production_budget_usd']
movies_data['domestic_roi'] = movies_data['domestic_gross_usd'] / movies_data['production_budget_usd']
movies_data['non_dom_roi'] = movies_data['non_dom_gross_usd'] / movies_data['production_budget_usd']
movies_data.head()
```

```python
movies_data['success_level_ww'] = movies_data['worldwide_roi'].apply(lambda x: 0 if x<1 else (1 if x<2 else 2))
```

```python
movies_data.info()
```

```python
movies_data.drop(['tagline', 'movie_title', 'release_date'], axis=1, inplace=True)
movies_data.dropna(inplace=True)
movies_data.head()
```

```python
movies_data.info()
```

```python
movies_data.describe()
```

```python
print(movies_data['rated'].unique()) # drops G
ohe = OneHotEncoder(drop='first', sparse=False)
movie_rating = movies_data[['rated']].copy()
ohe.fit(movie_rating)
movie_rating_ohe = pd.DataFrame(
    data=ohe.transform(movie_rating),
    columns=[f"rating_{rating}" for rating in ohe.categories_[0][1:]]
)
movie_rating_ohe
```

```python
print(movies_data['genre'].unique()) # drops Action
ohe = OneHotEncoder(drop='first', sparse=False)
movie_genre = movies_data[['genre']].copy()
ohe.fit(movie_genre)
movie_genre_ohe = pd.DataFrame(
    data=ohe.transform(movie_genre),
    columns=[f"genre_{genre}" for genre in ohe.categories_[0][1:]]
)
movie_genre_ohe
```

```python
movies_modeling_data = pd.concat([movies_data, movie_rating_ohe, movie_genre_ohe], axis=1)
movies_modeling_data.head()
```

```python
movies_modeling_data.drop(['rated', 'genre'], axis=1, inplace=True)
movies_modeling_data.head()
```
